```python
import subprocess
import sys
import pylab as pl
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
from sklearn import metrics

import pandas as pd
import matplotlib.pyplot as plt
import os

import sklearn as skl
import seaborn as sns
import matplotlib as mpl
from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.grid_search import GridSearchCV
%matplotlib inline
mpl.use('Agg')


import pylab as pl

# constants
CURRENT_DIR = os.path.dirname(os.path.realpath("__file__"))


with open(os.path.join(CURRENT_DIR, 'bank.csv')) as infile:
    bank_data = pd.read_csv(infile, sep=';', na_values=['?'])

bank_data = bank_data.dropna()
```

```python
# convert categorical variable into dummies
# a = pd.concat([bank_data, pd.get_dummies(bank_data['job']), pd.get_dummies(bank_data['marital'],pd.get_dummies(bank_data['education']))],axis=1)
bank_data['y']=bank_data['y'].replace(['yes', 'no'], [1, 0])
```

```python
# pd.concat([bank_data,pd.get_dummies(bank_data['job']),pd.],axis=1)
import numpy as np
categorical_columns = bank_data.loc[:, bank_data.dtypes == np.dtype('O')].columns.values
```

```python
bank_data_merge = pd.concat([pd.get_dummies(bank_data[categorical_columns]),bank_data],axis=1)
```

```python
bank_data_dummy = bank_data_merge.loc[:, bank_data_merge.dtypes != np.dtype('O')]
Y = bank_data_dummy.iloc[:,-1]
X = bank_data_dummy.ix[:,bank_data_dummy.columns != 'y']
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

# GBM model grid search

```python
hypergrid = {'max_depth': [2, 3],
             'max_features': [10, 15],
             'subsample': [.8, 1],
             'learning_rate': [0.01, 0.05]
            }

clf = GridSearchCV(GBC(verbose=1, n_estimators=200), hypergrid, scoring='roc_auc', verbose=1, cv=2, refit=True)
```

```python
clf.fit(X_train,Y_train)
```

```python
scores = pd.DataFrame(clf.grid_scores_)
scores.columns = ['parameters', 'scores', 'confidence_interval']
```

```python
scores.sort('scores', ascending=False).head(10)
```

```python
print(clf.best_params_)
print(clf.best_score_)
```

```python
def plot_roc(model, estimators, train_x, train_y, val_x, val_y):
    model.n_estimators = estimators
    model.fit(train_x, train_y)
    loss_vector = np.zeros(estimators)
    for ii, predicted in enumerate(model.staged_decision_function(val_x)):
        loss_vector[ii] = skl.metrics.roc_auc_score(val_y, predicted)
    plt.plot(loss_vector)
    plt.xlabel('Boosting iterations')
    plt.ylabel('Loss')
```

```python
best_hyperparams = scores.sort('scores', ascending=False).head(3)['parameters']
```

```python
for params in best_hyperparams:
    plot_roc(GBC(**params), 100, X_train,Y_train,X_test,Y_test)
plt.legend(best_hyperparams, bbox_to_anchor=(1, 0.2))
```

# Predict using test dataset

```python
best_gbm = GBC(**list(best_hyperparams)[0])
best_gbm.fit(X_train, Y_train)

second_gbm = GBC(**list(best_hyperparams)[1])
second_gbm.fit(X_train, Y_train)

third_gbm = GBC(**list(best_hyperparams)[2])
third_gbm.fit(X_train, Y_train)
```

# Somer's D and lift chart

```python
def get_somersd(data_list):
    '''
    Input to get_somersd is a list of tuples. Each tuple consists of a label,
    the actual target, and the predicted target.
    Actual and predicted targets
    are in Pandas series or 1-dimensional arrays or data frames.
    
    Returns a list of Somers' D values and prints the Somers' D by input data.
    '''
    somersd_list=[]
    for label, actual, predicted in data_list:
        sd=(skl.metrics.roc_auc_score(actual, predicted)-0.5)*2
        somersd_list.append(sd)
        print(label + " Somers' D: "+str(sd))
    return somersd_list
```

```python
input_list = [("best_gbm", Y_test, best_gbm.predict_proba(X_test)[:,1]),
              ("second_gbm", Y_test, second_gbm.predict_proba(X_test)[:,1]),
              ("third_gbm", Y_test, third_gbm.predict_proba(X_test)[:,1])]
somersds = get_somersd(input_list)
```

```python
def somersd(score, target, bins=500):
    pred = np.array([score, target]).T
    score, target = pred[pred.argsort(0)[:,0]].T
    #score, target = np.array(pd.DataFrame(pred).sort((0,1))).T
    
    n_1=float(sum(target))
    n_0=float(len(score)-n_1)
    
    pct_1=[0.0]
    pct_0=[0.0]

    for i in xrange(bins):
        end = int(float(i+1)/bins*len(pred))
        c_1 = sum(target[0:end])
        pct_1.append(c_1/n_1)
        pct_0.append((end-c_1)/n_0)

    pct_1.append(1.0)
    pct_0.append(1.0)
    
    sd=[(pct_1[i+1]+pct_1[i])*(pct_0[i+1]-pct_0[i]) for i in xrange(bins)]
    
    somersd=1-sum(sd)
    return somersd

def somers_plots(preds, bins=40):
    figsize(16,8)
    df = pd.DataFrame()
    max_yvalue=[]
    for model, pred in preds.iteritems():
        mname = str(model)[:str(model).find("(")]
        #p = np.array(pd.DataFrame(pred).sort((0,1)))
        p = pred[pred.argsort(0)[:,0]]
        pct = []
        for i in xrange(bins):
            start = int(float(i)/bins*len(p))
            end = int(float(i+1)/bins*len(p))
            pct.append((i, np.mean(p[start:end,1])))
        pct = np.array(pct)
        print pct
        for i in pct[:,1]:
            max_yvalue.append(i)
        som = ' (%.3f)' % somersd(p[:,0], p[:,1])
        df = pd.concat([df, pd.DataFrame([pct, mname+som, som]).T])

    for k, v in df.sort(2, ascending=False).iterrows():
        pct = v.get_values()[0]
        plt.plot(pct[:,0], pct[:,1], label=v[1])
    
    ylim(0,max(max_yvalue)+0.05)
    legend(loc='upper left')
    show()
```
