## Univariate Charts

````python
def midpoints(edges):
    return (edges[:-1] + edges[1:]) / 2

def univariate_plot(var, target, bins=50):
    dataset = pd.DataFrame({'var': var, 'target': target})

    binned, edges = pd.cut(dataset['var'], bins, retbins=True)
    grouped_data = dataset.groupby(binned)['target']
    target_means = grouped_data.mean().fillna(0)
    bin_centers = midpoints(edges)
            
    fig, ax1 = plt.subplots()
    ax1.yaxis.tick_right()
    ax1.plot(bin_centers, target_means, '-', label=var)

    ax2 = ax1.twinx()
    ax2.hist(dataset['var'], bins, alpha=0.2, histtype='step', fill=True)
    ax2.set_xlabel(var)
    ax2.set_ylabel('Counts')
    ax2.grid(b=False)
    plt.xlim([dataset['var'].min(), dataset['var'].max()])
    return fig, [ax1, ax2]
````

## Munging and Transforming

### Check categorical columns that might not actually be categorical
The suspicious columns have an abnormally large number of unique variables. Even though the datatype was 'Object', suggesting that it was a categorical, they might be something else. Closer inspection reveals they are all currencies stored as strings, so we should convert them to floats.
Be aware that if you use your own data, not all suspicious columns are actually continuous variables, so make sure to check this code accordingly.

````python
categorical_columns = build_data.loc[:, build_data.dtypes == dtype('O')].columns.values
category_counts = {col: len(build_data[col].unique()) for col in categorical_columns}

discrete_columns = []
for column_name, counts in category_counts.items():
    if counts < categorical_cutoff:
        discrete_columns.append(column_name)

suspicious_columns = set(categorical_columns) - set(discrete_columns)
for column in suspicious_columns:
    print(build_data[column].head(), '\n\n')
````

### Cleanup, imputation, cap and floor, dummy variables

Now we do some data cleaning - cap, floor, missing imputation, and dummy creation

These tasks are routine in data analysis, but can often be difficult to debug if things go wrong. One reason why is because the logic for each data cleaning task, while simple, can be detailed and implementation-specific. In the following block, we will practice what's called the "method extraction refactoring" method, where individual, independent tasks are extracted from a more complicated function.

The important function here is `clean_up_dataset()` which cleans input data according to some parameters. However, this function performs many tasks, and each task is extracted into it's own method. This kind of refactoring has several advantages; beyond improving the reusability of each piece of functionality, mechanical operations that fall under a clean conceptual umbrella can be properly renamed. Leading to a final function that's not only readable, but debuggable even if you don't know what the rest of the methods do.

````python
def clean_up_dataset(df, target, cap=0.99, floor=0.01):
    unary_list = get_unary(df, target)
    discrete_list = get_discrete(df, target)
    unpacked_discrete = pd.get_dummies(df.ix[:, discrete_list])
    continuous_list = list(set(df.columns.values) - set(discrete_list + unary_list + [target]))
    capped_and_floored = cap_and_floor(df[continuous_list], cap, floor)
    cleaned = impute(capped_and_floored)
    return pd.concat([unpacked_discrete, cleaned, df[target]], axis=1)

def get_unary(df, target):
    value_counts = df.apply(lambda x: len(x.unique()))
    unary_list = list(value_counts.ix[value_counts == 1].index.values)
    if target_column in unary_list:
        unary_list.remove(target_column)
    return unary_list

def get_discrete(df, target):
    unary_list = get_unary(df,target)
    discrete_list = list(df.ix[:, df.dtypes == dtype('O')].columns.values)
    for column in discrete_list:
        if column in unary_list:
            discrete_list.remove(column)
    return discrete_list

def cap_and_floor(df, cap=0.99, floor=0.01):
    new_df=pd.DataFrame()
    for column in df.columns.values:
        new_df[column]=df[column].clip(df[column].quantile(floor), df[column].quantile(cap))
    return new_df

def impute(df):
    return df.fillna(df.median())
````

````python
build_data_cleaned = clean_up_dataset(build_data,target_column)
intime_data_cleaned = clean_up_dataset(intime_data, target_column)
outoftime_data_cleaned = clean_up_dataset(outoftime_data, target_column)
````

The data cleaner created dummy indicators for all levels of each categorical variable, including ones that do not appear in all three datasets. 

````python
def find_common_columns(*dfs):
    return list(set.intersection(*[set(df.columns) for df in dfs]))
````

````python
common_columns = find_common_columns(build_data_cleaned, intime_data_cleaned, outoftime_data_cleaned)
build_data_cleaned = build_data_cleaned.ix[:, common_columns]
intime_data_cleaned = intime_data_cleaned.ix[:, common_columns]
outoftime_data_cleaned = outoftime_data_cleaned.ix[:, common_columns]
````

````python
build_Y = build_data_cleaned[target_column]
intime_Y = intime_data_cleaned[target_column]
outoftime_Y = outoftime_data_cleaned[target_column]
build_X = build_data_cleaned.drop([target_column], axis=1)
intime_X = intime_data_cleaned.drop([target_column], axis=1)
outoftime_X = outoftime_data_cleaned.drop([target_column], axis=1)
````

## Parameter Calibration
Now we are ready to do some modeling. Suppose we want to build a gradient boosting classifier.
That type of model has a lot of different parameters, and the best way to figure out which parameters to use is to try them. To use SKLearn's built-in gridsearch function, define a dictionary with the parameters you want to vary as keys, and lists of the values for those parameters that you want to test as values. The function will then build a model for every combination of those parameter values.

````python
hypergrid = {'max_depth': [2, 3],
             'max_features': [20, 30],
             'subsample': [.5, 1],
             'learning_rate': [0.05, 0.1]
            }

clf = GridSearchCV(GBC(verbose=1, n_estimators=200), hypergrid, scoring='roc_auc', verbose=1, cv=2, refit=True)
````

The `fit()` method will run the grid search over the hyperparameter space.

````python
clf.fit(build_X, build_Y)
````

The after the search is run, the GridSearchCV object stores information about each of the runs. When cross-validation is used, a confidence interval is also given for each set of parameters.

````python
scores = pd.DataFrame(clf.grid_scores_)
scores.columns = ['parameters', 'scores', 'confidence_interval']
````

````python
scores.sort('scores', ascending=False).head(10)
````

Gridsearch also stores the best parameter setup and score.

````python
print(clf.best_params_)
print(clf.best_score_)
````

## Calibrating the number of trees

````python
def plot_roc(model, estimators, train_x, train_y, val_x, val_y):
    model.n_estimators = estimators
    model.fit(train_x, train_y)
    loss_vector = np.zeros(estimators)
    for ii, predicted in enumerate(model.staged_decision_function(val_x)):
        loss_vector[ii] = skl.metrics.roc_auc_score(val_y, predicted)
    plt.plot(loss_vector)
    plt.xlabel('Boosting iterations')
    plt.ylabel('Loss')
````

````python
best_hyperparams = scores.sort('scores', ascending=False).head(3)['parameters']
````

````python
for params in best_hyperparams:
    plot_roc(GBC(**params), 100, build_X, build_Y, intime_X, intime_Y)
plt.legend(best_hyperparams, bbox_to_anchor=(1, 0.2))
````

## Somers' D and Lift Chart

````python
def get_somersd(data_list):
    '''
    Input to get_somersd is a list of tuples. Each tuple consists of a label,
    the actual target, and the predicted target.
    Actual and predicted targets
    are in Pandas series or 1-dimensional arrays or data frames.
    
    Returns a list of Somers' D values and prints the Somers' D by input data.
    '''
    somersd_list=[]
    for label, actual, predicted in data_list:
        sd=(skl.metrics.roc_auc_score(actual, predicted)-0.5)*2
        somersd_list.append(sd)
        print(label + " Somers' D: "+str(sd))
    return somersd_list
````

````python
def plot_lift_chart(score, target, quantiles=10):
#    fig, ax = plt.subplots()
    lift = pd.concat([pd.Series(target), pd.Series(score)], axis=1)
    lift.columns = ['Actual','Prediction']
    buckets = pd.qcut(lift['Prediction'], quantiles, labels=False)
    lift['Bucket'] = buckets
    liftchart = pd.DataFrame()
    liftchart['Risk'] = lift.groupby(['Bucket'])['Actual'].mean()
    liftchart['Bucket'] = pd.Series(range(quantiles))
    plt.plot(liftchart['Bucket'], liftchart['Risk'])
    xlabels = liftchart.index.tolist()
#    ax.plot(liftchart['Bucket'], liftchart['Risk'], 'o-')
#    ax.set_xticks(liftchart['Bucket'])
#    ax.set_xticklabels(xlabels)

#    if predictor_name is not None:
#        ax.set_title('Lift chart for {0}'.format(predictor_name))

#    return fig, ax
````

````python
for label, target, score in input_list:
    plot_lift_chart(score, target, quantiles=10)
plt.xlabel('Model Bucket')
plt.ylabel('Diabolical Risk')
plt.legend([label for label, target, score in input_list], bbox_to_anchor=(0.25, 1))
sns.despine()
````
