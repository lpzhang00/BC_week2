```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
from sklearn import metrics

import pandas as pd
import matplotlib.pyplot as plt
import os

# constants
CURRENT_DIR = os.path.dirname(os.path.realpath("__file__"))
DATA_DIR = os.path.abspath(os.path.join(CURRENT_DIR, 'data'))
BANK_DIR = os.path.join(DATA_DIR, 'bank')

with open(os.path.join(BANK_DIR, 'bank.csv')) as infile:
    bank_data = pd.read_csv(infile, sep=';', na_values=['?'])

bank_data = bank_data.dropna()
```

```python
# convert categorical variable into dummies
# a = pd.concat([bank_data, pd.get_dummies(bank_data['job']), pd.get_dummies(bank_data['marital'],pd.get_dummies(bank_data['education']))],axis=1)
bank_data['y']=bank_data['y'].replace(['yes', 'no'], [1, 0])
```

```python
# pd.concat([bank_data,pd.get_dummies(bank_data['job']),pd.],axis=1)
import numpy as np
categorical_columns = bank_data.loc[:, bank_data.dtypes == np.dtype('O')].columns.values
```

```python
bank_data_merge = pd.concat([pd.get_dummies(bank_data[categorical_columns]),bank_data],axis=1)
```

```python
bank_data_dummy = bank_data_merge.loc[:, bank_data_merge.dtypes != np.dtype('O')]
Y = bank_data_dummy.iloc[:,-1]
X = bank_data_dummy.ix[:,bank_data.columns != 'y']
```

```python
bank_data_dummy[]
```

```python
acc = []
pre = []
rec = []
f1 = []

for k in range(1,10):
    x = KNeighborsClassifier(k)
    results_1 = cross_val_score(x, X, Y, cv=5)
    results_2 = cross_val_score(x, X, Y, cv=5, scoring='precision_macro')
    results_3 = cross_val_score(x, X, Y, cv=5, scoring='recall_macro')
    results_4 = cross_val_score(x, X, Y, cv=5, scoring='f1_macro')
    acc.append((k, results_1.mean()))
    pre.append((k, results_2.mean()))
    rec.append((k, results_3.mean()))
    f1.append((k, results_4.mean()))

top_acc = sorted(acc, key=lambda p: p[1], reverse=True)
top_pre = sorted(pre, key=lambda p: p[1], reverse=True)
top_rec = sorted(rec, key=lambda p: p[1], reverse=True)
top_f1 = sorted(f1, key=lambda p: p[1], reverse=True)
```
